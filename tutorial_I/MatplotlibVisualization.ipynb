# Matplotlib Visualization 

In this script, we will learn how to **create table** and get to know more about **different kinds of graphs to perform data visualization using Matplotlib**. :)

First, let us take a look at how to create a table.   

```python
# import pandas and matplotlib 
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
# create 2D array of our example table given in the slides  
data = [['E001', 'M', 34, 123, 'Normal', 350], 
        ['E002', 'F', 40, 114, 'Overweight', 450], 
        ['E003', 'F', 37, 135, 'Obesity', 169], 
        ['E004', 'M', 30, 139, 'Underweight', 189], 
        ['E005', 'F', 44, 117, 'Underweight', 183], 
        ['E006', 'M', 36, 121, 'Normal', 80], 
        ['E007', 'M', 32, 133, 'Obesity', 166], 
        ['E008', 'F', 26, 140, 'Normal', 120], 
        ['E009', 'M', 32, 133, 'Normal', 75], 
        ['E010', 'M', 36, 133, 'Underweight', 40] ] 
```

```python
# convert the above 2D array into a dataframe
# so that we can analyze and visualize these data better
df = pd.DataFrame(data, columns = ['EMPID', 'Gender',  
                                    'Age', 'Sales', 
                                    'BMI', 'Income'] ) 
# df = pandas.Dataframe(...)
```

```python
df
```

Next, let us dive into the colorful world of graphs!

**Histogram**

In below code histogram is plotted for *Age*, *Income*, and *Sales*. So these plots in the output shows frequency of each unique value for each of the three attributes.

```python
# create histogram for numeric data 
df.hist() 
  
# show plot 
plt.show() 
```

**Bar Plot** 

```python
# Dataframe of previous code is used here 
  
# Plot the bar chart for numeric values 
# a comparison will be shown between 
# all three attributes: age, income, sales 
df.plot.bar() 
plt.xlabel("EMPID Index")
plt.ylabel("Value")
```

```python
# plot between 2 attributes 
plt.bar(df['Age'], df['Sales']) #call the function
plt.xlabel("Age")  # give xlables and ylables
plt.ylabel("Sales") 
plt.show() #show the graph
```

**Box Plot**

```python
# For each numeric attribute of dataframe 
df.plot.box() 
```

```python
# individual attribute box plot 
plt.boxplot(df['Income']) 
plt.xlabel("Income") 
plt.show() 
```

**Scatter plot**

```python
# scatter plot between income and age 
plt.scatter(df['Income'], df['Age']) 
plt.xlabel("Income") 
plt.ylabel("Age") 
plt.show() 
```

```python
# scatter plot between income and sales 
plt.scatter(df['Income'], df['Sales']) 
plt.xlabel("Income") 
plt.ylabel("Sales")
plt.show()
```

```python
# scatter plot between sales and age 
plt.scatter(df['Sales'], df['Age']) 
plt.xlabel("Sales") 
plt.ylabel("Age")
plt.show()
```

**Pie Chart**

```python
# pie chart for Age
plt.pie(df['Age'], labels = {"A", "B", "C", 
                             "D", "E", "F", 
                             "G", "H", "I", "J"}, 
                               
autopct ='% 1.1f %%', shadow = True) 
plt.show() 
```

```python
# pie chart for Income
plt.pie(df['Income'], labels = {"A", "B", "C", 
                                "D", "E", "F", 
                                "G", "H", "I", "J"}, 
                                  
autopct ='% 1.1f %%', shadow = True) 
plt.show() 
```

```python
# pie chart for Sales
plt.pie(df['Sales'], labels = {"A", "B", "C", 
                               "D", "E", "F", 
                               "G", "H", "I", "J"}, 
autopct ='% 1.1f %%', shadow = True) 
plt.show() 
```
